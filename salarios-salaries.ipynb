```python
import pandas as pd
import numpy as np
import matplotlib as mp
import seaborn as sbn
```

```python
def importDataset(arquivo):
    try:
        global dados
        dados = pd.read_csv(arquivo)
        
    except NameError:
        print("Erro ao realizar upload dos dados! \n Operation Failed!")
    else:
        print("Upload realizado com sucesso! \n Upload successfully!")
        return dados
```

* ***VERIFICANDO OS DADOS - BR***
* ***VERIFYING DATASET - EN***

```python
 #-> Verificando quantos registros possuímos no dataset
#-> Verifying how much fields and columns we've in dataset

def countShape(dataframe):
    try:
        dados = dataframe.shape
    except NameError:
        return print("Erro ao buscar quantidade de colunas e registros \n Operation Failed!")
    else:
        return print("Quantidade de registros e colunas: \n Number of records and columns: ", dados)
```

* ***Verificando o cabeçalho e exibindo os 5 primeiros registros!***
* ***Analyzing head of dataset and displaying the 5th database records***

```python
#-> Verificando quantos registros possuímos no dataset
#-> Verifying how much fields and columns we've in dataset
def headDataset(dataframe, nIndice):
    try:
        dados = dataframe.head(nIndice)
    except NameError:
        print("Erro ao buscar cabeçalho do dataset! \n Error to load head of dataset!")
    else:
        return dados
```

* ***verificando campos nulos/em branco***
* ***Analyzing blank/null fields***

```python
#-> Verificando quantos registros possuímos no dataset
#-> Verifying how much fields and columns we've in dataset

def naDataset(dataframe):
    try:
        dados = dataframe.isna().sum()
    except NameError:
        return print("Erro ao buscar campos vázios \n Operation Failed!")
    else:
        print("Total de campos vázios: \n Total blank fields: ")
        return dados
```

* ***Verificando quais são tipos dos campos no dataset***
* ***Verifying which type fields of dataset***

```python
#-> Verificando quantos registros possuímos no dataset
#-> Verifying how much fields and columns we've in dataset

def nullDataset(dataframe):
    try:
        dados = dataframe.isnull().sum()
    except NameError:
        return print("Erro ao somar valores nulos \n Operation Failed!")
    else:
        print("Total de valores nulos: \n Total null fields: ")
        return dados
```

* ***Excluindo caracteres especiais em campos numéricos***
* ***Cleaning special caracters in numeric fields***

* ***Vou utilizar uma função para realizar esta conversão***
* ***I'll use function to realize that conversion***

```python
#-> Verificando quantos registros possuímos no dataset
#-> Verifying how much fields and columns we've in dataset

def cleanSpecial(dataframe, colunas):
    char = ["<","=",">","k","K"," "]    
    
    try:
        for c in char :
            dataframe[colunas] = dataframe[colunas].str.replace(c,"", n = 1)
            data = dataframe[colunas]
            
    except NameError:
        return print("Essa operação falhou! \n Operation Failed")
    
    else:
        print("Operação realizada com êxito!")
        return data
```

* ***Vamos precisar converter alguns campos do tipo object para o tipo float***
* ***We'll need to convert some fields of object type to float type***

```python
#-> Verificando quantos registros possuímos no dataset
#-> Verifying how much fields and columns we've in dataset

def objToflt(dataframe, colunas):
    try:
        if dataframe is not None:
            dataframe[colunas] = dataframe[colunas].astype(int)
            
    except NameError:
        return print("Error ao trocar objcect para float! \n Something wrong to change type of field")
    
    else:
        print("Campo atualizado com sucesso! \n Fields of dataset changed successfully")
        return dataframe[colunas]
```
